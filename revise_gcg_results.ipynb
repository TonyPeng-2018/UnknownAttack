```python
# revise the the best suffix of data
for key in data:
    # find the minimal loss in the loss history
    loss_history = data[key]["loss_history"]
    min_loss_idx = loss_history.index(min(loss_history))
    data[key]["ids_suffix_best"] = data[key]["ids_suffix_history"][min_loss_idx]
    data[key]["best_loss"] = loss_history[min_loss_idx]
```
